## 1. Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

## 2. Load the prepared train/test data

```python
import pandas as pd

X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")
y_train = pd.read_csv("y_train.csv")
y_test = pd.read_csv("y_test.csv")

y_train = y_train["Churn"]
y_test = y_test["Churn"]

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## 3. Quick look at the data

```python
X_train.head()
```

## 4. Class balance in the training set

```python
y_train.value_counts()
```

## 5. Train the baseline model – Logistic Regression

This is the required **first model** for classification.


```python
log_model = LogisticRegression(max_iter=2000)
log_model.fit(X_train, y_train)

log_pred = log_model.predict(X_test)

log_accuracy = accuracy_score(y_test, log_pred)
log_precision = precision_score(y_test, log_pred)
log_recall = recall_score(y_test, log_pred)
log_f1 = f1_score(y_test, log_pred)

print("Logistic Regression results")
print("Accuracy :", round(log_accuracy, 4))
print("Precision:", round(log_precision, 4))
print("Recall   :", round(log_recall, 4))
print("F1-score :", round(log_f1, 4))

print("\nClassification Report:")
print(classification_report(y_test, log_pred))
```

## 6. Train Model 2 – KNN

KNN is one of the algorithms we studied


```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

knn_pred = knn_model.predict(X_test)

knn_accuracy = accuracy_score(y_test, knn_pred)
knn_precision = precision_score(y_test, knn_pred)
knn_recall = recall_score(y_test, knn_pred)
knn_f1 = f1_score(y_test, knn_pred)

print("KNN results")
print("Accuracy :", round(knn_accuracy, 4))
print("Precision:", round(knn_precision, 4))
print("Recall   :", round(knn_recall, 4))
print("F1-score :", round(knn_f1, 4))

print("\nClassification Report:")
print(classification_report(y_test, knn_pred))
```

## 7. Train Model 3 – Decision Tree



```python
tree_model = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_model.fit(X_train, y_train)

tree_pred = tree_model.predict(X_test)

tree_accuracy = accuracy_score(y_test, tree_pred)
tree_precision = precision_score(y_test, tree_pred)
tree_recall = recall_score(y_test, tree_pred)
tree_f1 = f1_score(y_test, tree_pred)

print("Decision Tree results")
print("Accuracy :", round(tree_accuracy, 4))
print("Precision:", round(tree_precision, 4))
print("Recall   :", round(tree_recall, 4))
print("F1-score :", round(tree_f1, 4))

print("\nClassification Report:")
print(classification_report(y_test, tree_pred))
```

## 8. 5-fold Cross-Validation

```python
log_cv = cross_val_score(log_model, X_train, y_train, cv=5, scoring="accuracy")
knn_cv = cross_val_score(knn_model, X_train, y_train, cv=5, scoring="accuracy")
tree_cv = cross_val_score(tree_model, X_train, y_train, cv=5, scoring="accuracy")

print("Logistic Regression CV scores:", np.round(log_cv, 4))
print("Mean CV accuracy:", round(log_cv.mean(), 4))
print("Std CV accuracy :", round(log_cv.std(), 4))
print()

print("KNN CV scores:", np.round(knn_cv, 4))
print("Mean CV accuracy:", round(knn_cv.mean(), 4))
print("Std CV accuracy :", round(knn_cv.std(), 4))
print()

print("Decision Tree CV scores:", np.round(tree_cv, 4))
print("Mean CV accuracy:", round(tree_cv.mean(), 4))
print("Std CV accuracy :", round(tree_cv.std(), 4))
```

## 9. Comparison table

```python
results = pd.DataFrame({
    "Model": ["Logistic Regression", "KNN", "Decision Tree"],
    "Accuracy": [log_accuracy, knn_accuracy, tree_accuracy],
    "Precision": [log_precision, knn_precision, tree_precision],
    "Recall": [log_recall, knn_recall, tree_recall],
    "F1-score": [log_f1, knn_f1, tree_f1],
    "CV Mean Accuracy": [log_cv.mean(), knn_cv.mean(), tree_cv.mean()],
    "CV Std Accuracy": [log_cv.std(), knn_cv.std(), tree_cv.std()]
})

results = results.round(4)
results
```

## 10. Simple visualization of model accuracy

```python
plt.figure(figsize=(8, 4))
plt.bar(results["Model"], results["Accuracy"])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.xticks(rotation=15)
plt.show()
```

## 11. Written interpretation

From the comparison table, **Logistic Regression** gives the best overall performance on this dataset.  
It has the highest test accuracy and the highest F1-score, so it gives the most balanced result among the three models.

**KNN** works, but its accuracy is lower than the other two models.  
**Decision Tree** gets a good accuracy, but its recall is lower, which means it misses more churn cases.

Overall, Logistic Regression is the best choice for this task because it is simple, stable, and performs best on the prepared Telco churn data.
